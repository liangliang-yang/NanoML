# One sample

* https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_1samp.html


```python
from scipy import stats
rvs = stats.norm.rvs(loc=5, scale=10, size=100)

stats.ttest_1samp(rvs, 4.0)
```
