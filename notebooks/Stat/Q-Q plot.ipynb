Q-Q plots are used to find the type of distribution for a random variable whether it be a Gaussian Distribution, Uniform Distribution, Exponential Distribution or even Pareto Distribution

* https://numpy.org/doc/stable/reference/random/generated/numpy.random.binomial.html

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

n = 2000
# np.random.binomial(n, p, 1000)
observation = np.random.binomial(n, 0.75, size = 1000)/n

# standardize the observation
z = (observation-np.mean(observation))/np.std(observation)

stats.probplot(z, dist="norm", plot=plt)
plt.title("Normal Q-Q plot")
plt.show()
```

```python
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')

plt.show()
```
