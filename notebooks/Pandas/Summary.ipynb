### groupby agg
* https://stackoverflow.com/questions/33780794/how-to-sum-negative-and-positive-values-separately-when-using-groupby-in-pandas

```python
df = pd.DataFrame({'A' : ['foo', 'bar', 'foo', 'bar', 'foo', 'bar', 'foo', 'foo'],
                   'B' : ['one', 'one', 'two', 'three', 'two', 'two', 'one', 'three'],
                   'C' : np.random.randn(8), 'D' : np.random.randn(8)})

df
```

```python
df.groupby(df['A'])['C'].agg(
    [('negative' , lambda x : x[x < 0].sum()) , 
     ('positive' , lambda x : x[x > 0].sum())]
)
```

### pandas get rows which are NOT in other dataframe
* https://stackoverflow.com/questions/28901683/pandas-get-rows-which-are-not-in-other-dataframe

```python
df1 = pd.DataFrame(data = {'col1' : [1, 2, 3, 4, 5], 'col2' : [10, 11, 12, 13, 14]}) 
df2 = pd.DataFrame(data = {'col1' : [1, 2, 3], 'col2' : [10, 11, 12]})
```

```python
# method 1
common = df1.merge(df2,on=['col1','col2'])
print(common)
df1[(~df1.col1.isin(common.col1))&(~df1.col2.isin(common.col2))]
```

```python
# method 2
df1[~df1.isin(df2)].dropna()
```
